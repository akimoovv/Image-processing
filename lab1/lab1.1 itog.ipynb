```python
from skimage.io import imread, imshow, imsave
from skimage import img_as_float, img_as_ubyte, img_as_uint
import numpy as np
```

```python
img = imread('A2.png')
img_greyscale = img[:,:,0]

imshow(img_greyscale)
```

```python
def encode(img,n):
    return (img/n).astype('uint8')
```

```python
def decode(img,n):  
    return img_as_ubyte(np.clip(img * n, 0, 255))
```

```python
img_greyscale_encode = encode(img_greyscale,8)
img_greyscale_decode = decode(img_greyscale_encode,8)
```

```python
imshow(img_greyscale_encode)
```

```python
imshow(img_greyscale_decode)
```

```python
imsave('img_greyscale_encode.png',img_greyscale_encode)
```
